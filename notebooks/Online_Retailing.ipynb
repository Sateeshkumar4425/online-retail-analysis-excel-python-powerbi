```python
data['Description'] = data['Description'].astype(str).str.replace(r'[()/] ?', '', regex=True).str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
print(f"Initial number of rows: {len(data)}")
# Drop rows where 'UnitPrice' is negative
data = data[data['UnitPrice'] >= 0]
print(f"Number of rows after dropping negative UnitPrice: {len(data)}")
# Drop rows where 'Description' is 'NaN' (originally null) or an empty string after cleaning
data = data[~((data['Description'] == 'NaN') | (data['Description'] == ''))]
print(f"Number of rows after dropping problematic Descriptions: {len(data)}")
```

```python
data.describe()
```

```python
print(f"Number of rows before applying new filter: {len(data)}")
# Condition 1: Description contains '?' or '1'
condition_description = data['Description'].str.contains(r'[?]', regex=True)
# Condition 2: UnitPrice is negative
condition_unitprice_negative = data['UnitPrice'] < 0
# Combined condition: (Description contains '?' or '1') AND (UnitPrice is negative)
rows_to_drop = condition_description & condition_unitprice_negative
# Drop rows that satisfy the combined condition
data = data[~rows_to_drop]
print(f"Number of rows after applying new filter: {len(data)}")
```

```python
orders_by_country = data.groupby('Country')['InvoiceNo'].nunique()

top5 = orders_by_country.sort_values(ascending=False).head(5)

others = orders_by_country.sum() - top5.sum()

pie_data = pd.concat([top5, pd.Series({'Others': others})])

```

```python
explode = [0.1 if value < pie_data.max() * 0.1 else 0 for value in pie_data]
plt.figure(figsize=(7,7))
plt.pie(
    pie_data,
    labels=pie_data.index,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode
)
plt.title('Orders Share by Country')
plt.tight_layout()
plt.show()
```

The pie chart shows that the United Kingdom dominates the total number of orders, contributing over 90% of all orders. All other countries together make up only a small share, indicating that most customer activity comes from the UK.

```python
top_countries = data['Country'].value_counts().head(5).index
box_data = data[data['Country'].isin(top_countries) & (data['UnitPrice_INR'] > 0)]
plt.figure(figsize=(10,6))
sns.boxplot(data=box_data, x='Country', y='UnitPrice_INR')
plt.yscale('log')
plt.title('Unit Price Distribution by Country')
plt.tight_layout()
plt.show()
```

The box plot shows that most product prices are similar across countries, but the United Kingdom has more extreme high-price outliers. Overall, prices are mostly concentrated at lower values, with a few very expensive items in every country.

```python
numeric_data = data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

The heatmap shows that UnitPrice and UnitPrice_INR are perfectly correlated, which confirms that the currency conversion from unit price to INR is accurate. Quantity has almost no correlation with price, indicating that the number of items purchased does not strongly depend on the product price.

```python
sns.pairplot(data=plot_data, hue='Country')
plt.suptitle('Pairplot of Unit Price by Country', y=1.02)
plt.show()
```

The chart shows that most products are sold at lower prices, while only a few items are very expensive, mainly from the United Kingdom. The prices in INR closely follow the original unit prices, which confirms that the currency conversion is correct. It also becomes clear that customers usually buy more quantity when the price is low, whereas expensive products are bought less often, regardless of the country.
